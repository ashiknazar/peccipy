```python
def hello():
    print("hello peccipy")

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

def impute_and_onehot_encode(df, columns_to_process):
    # Make a copy of the original dataframe
    df_processed = df.copy()
    
    # Iterate through each column in columns_to_process
    for col in columns_to_process:
        # Impute missing values with most frequent value
        imputer = SimpleImputer(strategy='most_frequent')
        df_processed[col] = imputer.fit_transform(df[[col]])
        
        # Perform one-hot encoding on the column
        encoder = OneHotEncoder(sparse=False, drop='first')
        encoded_cols = pd.DataFrame(encoder.fit_transform(df_processed[[col]]),
                                    columns=encoder.get_feature_names([col]))
        
        # Concatenate the encoded columns back to the dataframe
        df_processed = pd.concat([df_processed, encoded_cols], axis=1)
        
        # Drop the original column after encoding
        df_processed.drop(col, axis=1, inplace=True)
    
    return df_processed


columns_to_process = ['Category']
df=pd.read_csv('titanic.csv')
print(df.head())
#processed_df = impute_and_onehot_encode(df, columns_to_process)
#print(processed_df)
```
